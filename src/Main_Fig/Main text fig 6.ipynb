```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

tcga_freq = pd.read_csv("C:\\Users\\HP\\Downloads\\tcga_freq_data.csv", index_col=(0))
df = tcga_freq.drop(columns='Cancer')

corr_df = df.corr()
corr_df = corr_df[['Pluripotent', 'Multipotent', 'Unipotent']]
corr_df = corr_df[~corr_df.index.isin(['Pluripotent', 'Multipotent', 'Unipotent', 'Endothelial', 'Malignant', 'Fibroblast', 'Epithelial'])]
corr_df = corr_df.sort_values(by='Pluripotent', ascending=False).T

from matplotlib.colors import CSS4_COLORS
# available_colors = list(CSS4_COLORS.values())
available_colors = [
    '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
    '#9b59b6', '#ff6347', '#2e8b57', '#add8e6', '#f0e68c', '#20b2aa', '#f4a460', '#ff1493', '#800080', '#ffd700',
    '#ff4500', '#adff2f', '#00bfff', '#ff1493', '#8b008b', '#800000', '#a52a2a', '#ffff00', '#32cd32', '#ff8c00'
]
column_colors = {col: color for col, color in zip(corr_df.columns, available_colors[:corr_df.shape[1]])}

import matplotlib.patches as mpatches
import matplotlib.cm as cm
from matplotlib.colors import CSS4_COLORS
import random

plt.figure(figsize=(10, 2))
sns.heatmap(corr_df, cmap='Blues')

# colormap = cm.Purples
y_colors = ['firebrick', 'darkorange', 'gold']
available_colors = list(CSS4_COLORS.values())  # Get a list of distinct CSS4 colors
# random.shuffle(available_colors)  # Shuffle to ensure randomness
x_colors = available_colors[:corr_df.shape[1]]

ax = plt.gca()

for idx, color in enumerate(y_colors):
    ax.add_patch(mpatches.Rectangle((-0.5, idx), 0.5, 1, color=color, transform=ax.transData, clip_on=False))

for idx, col in enumerate(corr_df.columns):
    color = column_colors[col]
    ax.add_patch(mpatches.Rectangle((idx, -0.5), 1, 0.5, color=color, transform=ax.transData, clip_on=False))
    
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

x_legend_labels = corr_df.columns
x_patches = [mpatches.Patch(color=column_colors[label], label=label) for label in x_legend_labels]
# Create a custom legend for the y-axis
y_legend_labels = corr_df.index
y_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(y_colors, y_legend_labels)]

# Combine both legends
plt.legend(handles=x_patches + y_patches, loc='upper right', bbox_to_anchor=(1.5, 1), title='')

ax.set_xticks([])
ax.set_yticks([])
# plt.savefig('immune_stem_corr_tcga.png', dpi=1200, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
data = pd.read_csv('C:/Users/HP/Downloads/pluripotent65185.csv', index_col=0)

# Set up the Seaborn style for a visually appealing plot
sns.set(style="white")

# Create a step chart
plt.figure(figsize=(3, 4))
for patient_id in data.index:
    plt.step(data.columns, data.loc[patient_id], where='mid', label=patient_id, linewidth=2, alpha=0.8)

# Customize the plot
plt.xlabel("Treatment Stage", fontsize=16)
plt.ylabel("CSC Frequency", fontsize=16)
plt.title("pluripotent-like", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the data
file_path = r"C:\Users\HP\Downloads\Pluripotent_filtered.csv"
data = pd.read_csv(file_path)

# Melt the data to make it long-form for plotting
data_melted = data.melt(
    id_vars=['PatientID', 'Response'], 
    value_vars=['Pre', 'On'], 
    var_name='Dose', 
    value_name='Value'
)

# Sort doses to ensure Pre comes before On
data_melted['Dose'] = pd.Categorical(data_melted['Dose'], categories=['Pre', 'On'], ordered=True)

# Create the step plot
g = sns.FacetGrid(data_melted, col="Response", sharey=False, height=3, aspect=0.8, palette="tab10")
g.map_dataframe(
    sns.lineplot, 
    x="Dose", 
    y="Value", 
    hue="PatientID", 
    style="PatientID", 
    markers=True, 
    dashes=False
)

# Customize the plot
g.set_axis_labels("Dose", "CSC frequency")
g.set_titles("{col_name}", fontsize=18)
g.add_legend(title="Patient ID")

# Adjust tick and label sizes
for ax in g.axes.flat:
    ax.tick_params(axis='both', which='major', labelsize=18)  # Adjust tick label size
    ax.xaxis.label.set_size(18)  # X-axis label size
    ax.yaxis.label.set_size(18)  # Y-axis label size

# Adjust spacing
g.tight_layout()

# Show the plot
plt.show()
```
