```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

signature_matrix = pd.read_csv("C:\\Users\\HP\\Desktop\\shruti di comp\\TCGA sig matrices\\sig_brain.csv", index_col = 0)
predicted_gep = pd.read_csv("C:\\Users\\HP\\Desktop\\shruti di comp\\TCGA gep preds\\Brain\\gep_Brain_preds_unknown.csv", index_col = 0)

# Create DataFrames
df_predicted = pd.DataFrame(predicted_gep)
df_signature = pd.DataFrame(signature_matrix)

# Combine the two matrices horizontally
combined_matrix = pd.concat([df_predicted, df_signature], axis=1)

# Compute the correlation matrix
correlation_matrix = combined_matrix.corr()

n=correlation_matrix.shape[0]
n1 = n/2
n1 = int(n1)
# Extract the lower right 10x10 portion
lower_right_10x10 = correlation_matrix.iloc[-n1:, -n1:]


plt.figure(figsize=(12, 10))
sns.heatmap(lower_right_10x10, cmap='YlGnBu', center=0, annot=False)

# Set titles and labels
plt.title('Heatmap of Predicted GEP and Signature Matrix Correlation', fontsize=20)
plt.xlabel('Predicted GEP', fontsize=20)
plt.ylabel('scRNAseq Ground Truth', fontsize=20)

# Remove tick labels
plt.xticks([])  # Remove x tick labels
plt.yticks([])  # Remove y tick labels

plt.tight_layout(pad=2)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the median gene expression CSV file
file_path = 'C:/Users/HP/Downloads/median_gene_expression.csv'
df = pd.read_csv(file_path, index_col=0)

# Sort top 1000 genes for each potency state in descending order
pluripotent_top1000 = df['Pluripotent'].sort_values(ascending=False).head(1000)
multipotent_top1000 = df['Multipotent'].sort_values(ascending=False).head(1000)
unipotent_top1000 = df['Unipotent'].sort_values(ascending=False).head(1000)

# Combine the top 1000 genes from all three potency states
combined_top1000 = pd.concat([pluripotent_top1000, multipotent_top1000, unipotent_top1000], axis=1)

# Ensure that the combined DataFrame has unique rows (genes) by dropping duplicates
combined_top1000 = combined_top1000.loc[~combined_top1000.index.duplicated()]
sns.set(style="white")
# Create a heatmap for the top 1000 genes across the three potency states
plt.figure(figsize=(6, 6))
sns.heatmap(combined_top1000, cmap="magma")

# Add labels and title
plt.title('Top 1000 Genes in Pluripotent, Multipotent, and Unipotent States', fontsize=14)
plt.xlabel('Potency States')
plt.ylabel('Genes')
plt.yticks(fontsize=10)
# Save the plot at 1200 DPI

# Show the heatmap
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Load the data
file_path = 'C:/Users/HP/Downloads/GEP_PRECOG/gep_Brain_preds_unknown_precog_descending_genes.csv'
df = pd.read_csv(file_path)

# Set the gene names as the index for easier sorting and plotting
df.set_index('Gene', inplace=True)

# Sort the DataFrame based on the 'Pluripotent' column in descending order
df = df.sort_values(by='Pluripotent', ascending=False)

# Define genes to highlight with red blocks, removing duplicates
highlight_genes = list(set([
    'NES', 'CD44', 'PROM1', 'NANOG', 'SOX2', 'POU5F1', 'FUT4', 'ITGB6', 'SOX9', 'MKI67', 
    'ALDH1A1', 'ABCG2', 'CXCR4', 'EPCAM', 'CD34', 'LGR5', 'CD24', 'ITGB1', 'ALCAM', 'KLF4', 
    'SLC3A2', 'BMI1', 'VIM', 'MME', 'CD29', 'NT5E', 'THY1', 'ICAM1', 'ANPEP', 'SALL4', 
    'SLC22A3', 'MYC', 'ROR1', 'VCAM1', 'NGFR', 'ABCB5'
]))


# Define the colors for each cell type (Pluripotent, Multipotent, Unipotent)
x_colors = ['firebrick', 'darkorange', 'gold']  # Example colors: red, blue, green
x_labels = df.columns

# Create the heatmap without x-axis labels and ticks, add color for each gene in highlight list
fig, ax = plt.subplots(figsize=(3, 8))
sns.heatmap(df, cmap='YlGn', annot=False, cbar_kws={'label': 'Expression Level'},
            yticklabels=False, xticklabels=False, ax=ax)

# Remove x-axis ticks and labels
ax.set_xticks([])
ax.set_xlabel('')

# Remove y-axis label
ax.set_ylabel('')

# Add a red block next to genes in `highlight_genes`
for i, gene in enumerate(df.index):
    if gene in highlight_genes:
        ax.add_patch(mpatches.Rectangle((-0.25, i), 0.25, 40, color='purple', clip_on=False, transform=ax.transData))

# Add square color blocks above the heatmap for each cell type
for idx, color in enumerate(x_colors):
    ax.add_patch(mpatches.Rectangle((idx, len(df)), 1, 30, color=color, transform=ax.transData, clip_on=False))

# Create a custom legend for the color blocks
legend_handles = [mpatches.Patch(color=color, label=label) for color, label in zip(x_colors, x_labels)]
plt.legend(handles=legend_handles, title="Cell Types", bbox_to_anchor=(1.25, 1), loc='upper left', borderaxespad=0.)


# Display the plot
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the directory containing the tissue files and list them
directory = 'C:/Users/HP/Downloads/GEP_SIGNAL'
tissue_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Define a list of 30 distinct hex color codes for gene plots
distinct_colors = [
    '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
    '#9b59b6', '#ff6347', '#2e8b57', '#add8e6', '#f0e68c', '#20b2aa', '#f4a460', '#ff1493', '#800080', '#ffd700',
    '#ff4500', '#adff2f', '#00bfff', '#ff1493', '#8b008b', '#800000', '#a52a2a', '#ffff00', '#32cd32', '#ff8c00'
]

# Create a figure with subplots, adjusting figure width for better visualization
fig, axes = plt.subplots(1, len(tissue_files), figsize=(30, 6), sharey=True)

# Generate a color gradient for the background colors
background_colors = plt.cm.Purples(np.linspace(0, 0.2, len(tissue_files)))

# Loop through each file and create a plot in a single subplot for each tissue
for idx, tissue_file in enumerate(tissue_files):
    # Load the tissue-specific CSV file
    df = pd.read_csv(os.path.join(directory, tissue_file), index_col=0)

    # Calculate the expression differences between potency states
    diff_plur_to_multipotent = df['Multipotent'] - df['Pluripotent']
    diff_multipotent_to_unipotent = df['Unipotent'] - df['Multipotent']
    diff_plur_to_unipotent = df['Unipotent'] - df['Pluripotent']

    # Identify the top 10 genes showing the largest increase for each comparison
    top_10_plur_to_multipotent = diff_plur_to_multipotent.nlargest(10)
    top_10_multipotent_to_unipotent = diff_multipotent_to_unipotent.nlargest(10)
    top_10_plur_to_unipotent = diff_plur_to_unipotent.nlargest(10)

    # Combine the top 10 genes from each comparison into a single list
    top_genes = pd.concat([top_10_plur_to_multipotent]).index.unique()

    # Select the corresponding subplot and set the background color
    ax = axes[idx]
    ax.set_facecolor(background_colors[idx])

    # Plot each gene's expression across potency states with distinct colors
    for i, gene in enumerate(top_genes):
        ax.plot(df.columns, df.loc[gene], label=gene, color=distinct_colors[i % len(distinct_colors)], linewidth=3)

        # Annotate gene name near the end of each line
        y_position = df.loc[gene].iloc[-1]  # Position for last point of the line
        ax.text(df.columns[-1], y_position, gene, color=distinct_colors[i % len(distinct_colors)], 
                fontsize=20, verticalalignment='center', horizontalalignment='left')

    # Remove x-axis and x-tick labels
    ax.set_xticks([])
    ax.set_xticklabels([])
    ax.set_yticklabels([])

# Add a single y-axis label for the entire figure with increased font size
#fig.text('Gene Expression', va='center', rotation='vertical', fontsize=50)

# Adjust layout for readability and reduce spacing between plots
plt.subplots_adjust(wspace=1)  # Reduce space between subplots


# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
file_path = r"C:\Users\HP\Downloads\gsea_500_result.csv"
data = pd.read_csv(file_path)

# Filter data for cell_type == 'Unipotent'
filtered_data = data[data['cell_type'] == 'Unipotent']

# Prepare data for plotting
filtered_data['-log10(P-value)'] = -np.log10(filtered_data['P-value'])
filtered_data['log10(Odds Ratio)'] = np.log10(filtered_data['Odds Ratio'])

# Map Gene_set to unique colors
unique_gene_sets = filtered_data['Gene_set'].unique()
color_map = {gene_set: idx for idx, gene_set in enumerate(unique_gene_sets)}
filtered_data['Gene_set_color'] = filtered_data['Gene_set'].map(color_map)

# Create the scatter plot
plt.figure(figsize=(8, 4))
scatter = plt.scatter(
    filtered_data['log10(Odds Ratio)'],
    filtered_data['-log10(P-value)'],
    c=filtered_data['Gene_set_color'],
    cmap='viridis_r',
    alpha=0.3
)

# Add a color legend
handles = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.viridis_r(color_map[gene_set] / len(color_map)), markersize=8, label=gene_set)
    for gene_set in unique_gene_sets
]
plt.legend(handles=handles, title='Gene_set', loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0.)

# Add cutoff lines
plt.axhline(y=1.30, color='red', linestyle='dotted', label='P-value cutoff (-log10)')
plt.axvline(x=0.18, color='red', linestyle='dotted', label='Odds Ratio cutoff (log10)')

# Annotate specific pathways
# Replace the pathway names below with the ones you want to annotate
pathways_to_annotate = ['WNT/Calcium non-Canonical Signaling', 'P38 MAPK/MAPK14 Signaling', 'Aerobic Glycolysis WP4629']  # Example pathway names
for pathway in pathways_to_annotate:
    row = filtered_data[filtered_data['Term'] == pathway]
    if not row.empty:
        row = row.iloc[0]  # Get the first occurrence if multiple exist
        plt.annotate(
            row['Term'],
            (row['log10(Odds Ratio)'], row['-log10(P-value)']),
            textcoords="offset points",
            xytext=(6, 3),
            ha='left',
            fontsize=12,
            color='black'
        )

# Add labels and title
plt.xlabel('log10(Odds Ratio)', fontsize=16)
plt.ylabel('-log10(P-value)', fontsize=16)
plt.title('GSEA Results for Pluripotent Cell Type')

# Add grid for better readability
plt.grid(alpha=0.3)

# Save the plot at 1200 DPI

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# File path
file_path = r"C:\Users\HP\Downloads\filtered_gsea_quiescence.csv"

# Load the data
data = pd.read_csv(file_path)

# Sort data by 'Signaling' for grouping
data_sorted = data.sort_values(by='Signaling')
terms = data_sorted['Term']
signaling_groups = data_sorted['Signaling']
heatmap_data = data_sorted[['Pluripotent', 'Multipotent', 'Unipotent']]

# Create a color palette for signaling groups
unique_signaling = signaling_groups.unique()
palette = sns.color_palette("tab10", len(unique_signaling))
signaling_colors = dict(zip(unique_signaling, palette))
row_colors = signaling_groups.map(signaling_colors)

# Plot the heatmap
sns.set(style="white")
g = sns.clustermap(
    heatmap_data,
    cmap="BuPu_r",
    row_cluster=False,
    col_cluster=False,
    row_colors=row_colors,
    xticklabels=False,
    yticklabels=False,
    cbar_kws={'label': '-log10(P-value)'},
    figsize=(10, 10)
)

# Add a legend for signaling groups
legend_patches = [plt.matplotlib.patches.Patch(color=color, label=label)
                  for label, color in signaling_colors.items()]
plt.legend(
    handles=legend_patches,
    title="Signaling Groups",
    bbox_to_anchor=(2.4, 1.5),
    loc='upper left',
    borderaxespad=0
)


# Show the plot
plt.show()

```
