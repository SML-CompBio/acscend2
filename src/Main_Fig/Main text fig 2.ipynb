```python
import matplotlib.pyplot as plt
from matplotlib import cm

# Data
cell_counts = {'Adams et al': 1154, 'Micheli et al.': 2757, 'Wang et al.': 36981, 'Lau et al': 370, 'Yan et al': 124}

# Normalize the color values using viridis colormap
colors = cm.viridis([i / len(cell_counts) for i in range(len(cell_counts))])

# Adding an invisible wedge to create the half-donut effect
half_donut_values = list(cell_counts.values()) + [sum(cell_counts.values())]
colors = list(colors) + [(1, 1, 1, 0)]  # Add a fully transparent color for the invisible part

# Plotting the half-donut chart
fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts = ax.pie(
    half_donut_values,
    startangle=180,               # Start from the horizontal position
    counterclock=False,            # Rotate clockwise
    colors=colors,
    wedgeprops=dict(width=0.3)     # Width of the donut ring
)

# Create legend labels
legend_labels = [f"{stem_cell_type} (n={count})" for stem_cell_type, count in cell_counts.items()]

# Display the legend
plt.legend(wedges[:-1], legend_labels, title="Source data", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Title and display
plt.title('Number of Cells per Stem Cell Type', fontsize=16)
plt.tight_layout()

# Save the image
plt.savefig('C:/Users/HP/Downloads/half_donut_stem_cell_types.png', dpi=1200, bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

# Load data
df = pd.read_csv('C:/Users/HP/Downloads/data_overview.csv', index_col=0)

# Counting the number of studies per cancer type
studies_per_cancer = df['cancer'].value_counts()

# Summing the cell counts per cancer type
cell_counts_per_cancer = df.groupby('cancer').sum().sum(axis=1)

# Normalize the color values for consistent coloring
cancer_types = studies_per_cancer.index
colors = cm.turbo([i / len(cancer_types) for i in range(len(cancer_types))])

# Adding an invisible wedge to create half-donut effect for both plots
half_donut_studies = studies_per_cancer.tolist() + [sum(studies_per_cancer)]
half_donut_cells = cell_counts_per_cancer.tolist() + [sum(cell_counts_per_cancer)]
colors_studies = list(colors) + [(1, 1, 1, 0)]  # Transparent color for invisible part
colors_cells = list(colors) + [(1, 1, 1, 0)]    # Transparent color for invisible part

# Set up the figure and axis for two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Plot the half-donut for studies
wedges1, _ = ax1.pie(
    half_donut_studies,
    startangle=180,
    counterclock=False,
    colors=colors_studies,
    wedgeprops=dict(width=0.3)
)
ax1.set_title('Number of Studies per Cancer Type', fontsize=16)

# Plot the half-donut for cell counts
wedges2, _ = ax2.pie(
    half_donut_cells,
    startangle=180,
    counterclock=False,
    colors=colors_cells,
    wedgeprops=dict(width=0.3)
)
ax2.set_title('Number of Cells per Cancer Type', fontsize=16)

# Creating a shared legend
legend_labels = [
    f"{cancer} - Studies: {studies_per_cancer[cancer]}, Cells: {cell_counts_per_cancer[cancer]}"
    for cancer in cancer_types
]
plt.legend(
    wedges1[:-1], legend_labels,
    title="Cancer Types",
    loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)
)

# Adjust layout and save
plt.tight_layout()
plt.savefig('C:/Users/HP/Downloads/half_donut_comparison.png', dpi=1200, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

entropy = pd.read_csv("C:\\Users\\HP\\Downloads\\entropy_score.csv", index_col=(0))

from matplotlib.colors import LinearSegmentedColormap


# Create custom colormaps
low = (np.array([255, 244, 132]) / 255).tolist()  # Light yellow
high = (np.array([84, 236, 231]) / 255).tolist()  # Cyan
colors = [low, high]

# Generate a gradient colormap
cm_gradient = LinearSegmentedColormap.from_list("CustomGradient", colors, N=36)

# Example data

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))

# Define gradient background
num_labels = 12
for i in range(num_labels):
    color = cm_gradient(i / (num_labels - 1))  # Get color for each label
    ax.axvspan(i - 0.5, i + 0.5, color=color, alpha=0.3, zorder=0)


flierprops = dict(marker='.', markerfacecolor='white', markeredgecolor='white', markersize=5, linestyle='none')

sns.boxplot(data=entropy, x='file2', y='scent', hue='class', width=0.9, flierprops=flierprops,
            hue_order=['Unipotent', 'Multipotent', 'Pluripotent'], ax=ax)
sns.despine(right=True, top=True, offset=10, trim=True)
plt.xticks(rotation=90, fontsize=12)  # Increase x tick label size
plt.yticks(fontsize=18)               # Increase y tick label size

# Label axes with larger font size
plt.xlabel('CANCER', fontsize=18)
plt.ylabel('SCENT', fontsize=18)

# Move legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=16)
# plt.savefig('scentropy_score_img.png', dpi=1200, bbox_inches='tight')
```
