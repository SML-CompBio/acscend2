```python
import matplotlib.pyplot as plt

# Data
data = {
    "Pluripotent-like": [1356, 1462],
    "Multipotent-like": [15563, 9039, 5859, 486],
    "Bi/unipotent-like": [63, 1154, 2757]
}

# Calculate total cells for each state
state_totals = {state: sum(cells) for state, cells in data.items()}

# Extract states and corresponding totals
states = list(state_totals.keys())
totals = list(state_totals.values())

# Define colors for the states
colors = ["#40E0D0","#008080", "#7FFFD4"]  # Add colors of your choice

# Create the donut plot
plt.figure(figsize=(8, 8))
wedges, texts = plt.pie(
    totals,
    colors=colors,
    wedgeprops=dict(width=0.3, edgecolor='w'),
    startangle=90
)

# Add a legend with State names and cell counts
legend_labels = [
    f"{state} (n={total})" for state, total in zip(states, totals)
]
plt.legend(
    wedges,
    legend_labels,
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    title="States",
    fontsize=16
)

# Add a central white circle for the "donut" effect
plt.gca().add_artist(plt.Circle((0, 0), 0.6, color='white'))

# Add a title
plt.title("Distribution of Cells by State", fontsize=14)


# Save the plot as an image
output_path = r"C:\Users\HP\Downloads\States_donut.png"

# Display the plot
plt.tight_layout()
plt.show()
```
