```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the ground truth and prediction data
original_file = r"C:\Users\HP\Downloads\SYD67\SYD67_Original.csv"
prediction_file = r"C:\Users\HP\Downloads\SYD67\ACSCeND_SYD67.csv"

# Read the files
original_data = pd.read_csv(original_file, index_col=0)
prediction_data = pd.read_csv(prediction_file, index_col=0)

# Ensure common columns and align data
common_columns = original_data.columns.intersection(prediction_data.columns)
original_data = original_data[common_columns]
prediction_data = prediction_data[common_columns]

# Flatten the data for x-y comparison
x = original_data.values.flatten()
y = prediction_data.values.flatten()

# Remove NaN values
valid_indices = ~np.isnan(x) & ~np.isnan(y)
x = x[valid_indices]
y = y[valid_indices]

# Calculate Pearson's r and p-value
r, p = pearsonr(x, y)

# Create a DataFrame for visualization
plot_data = pd.DataFrame({
    "Ground Truth": x,
    "Prediction": y,
    "Column": np.repeat(common_columns, len(original_data))
})

# Set up the plot style
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(4, 4))

# Create scatter plot with points colored by column
palette = sns.color_palette("husl", len(common_columns))
sns.scatterplot(
    data=plot_data, x="Ground Truth", y="Prediction", hue="Column", palette=palette, s=50, alpha=0.6, edgecolor="None"
)

# Add regression line
sns.regplot(
    data=plot_data, x="Ground Truth", y="Prediction", scatter=False, 
    line_kws={"color": "black", "lw": 2, "label": "Regression Line"}
)

# Annotate with Pearson's r and p-value
plt.text(
    0.05, 0.95, f"Pearson's r = {r:.2f}\nP-value = {p:.2e}", 
    ha='left', va='top', transform=plt.gca().transAxes, fontsize=14,
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5')
)

# Final touches
plt.title("Scatter Plot of ACSCeND Predictions vs Ground Truth", fontsize=16)
plt.xlabel("SYD67 Ground Truth", fontsize=14)
plt.ylabel("ACSCeND Predictions", fontsize=14)
plt.legend(title="Column", bbox_to_anchor=(1.05, 1), loc='upper left')
sns.despine(right=True, top=True, offset=10, trim=True)
#output_path = r"C:\Users\HP\Downloads\SYD67_CORR.png"
#plt.savefig(output_path, dpi=1200, bbox_inches='tight')
plt.tight_layout()

# Show plot
plt.show()
```
