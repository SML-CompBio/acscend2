```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for the plot
tests = ['ACSCeND', 'TAPE', 'Scaden', 'CIBERSORTx', 'EPIC', 'DWLS', 'Bisque']
ccc_values = [0.95240,  0.8629, 0.7830, 0.7665, 0.5529, 0.7281, 0.7189]

# Define unique edge colors for each test/tool
edge_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# Calculate y-axis limits to emphasize differences
y_min = np.floor(min(ccc_values) * 1000) / 1000  # Rounding down to nearest 0.001
y_max = np.ceil(max(ccc_values) * 1000) / 1000  # Rounding up to nearest 0.001

# Plotting
sns.set(style="white")
plt.figure(figsize=(4, 4))

# Create barplot with individual edge colors
for i, (test, ccc, color) in enumerate(zip(tests, ccc_values, edge_colors)):
    plt.bar(i, ccc, edgecolor=color, color=color, linewidth=1.5)

# Adjust y-axis limits for better representation
plt.ylim((y_min-0.1), y_max)

# Adding labels, title, and customizing the style
plt.ylabel('CCC', fontsize=14)
plt.yticks(fontsize=14)
plt.title('PBMC10k independent Data', fontsize=16)

# Remove x-axis and x-tick labels
plt.xticks([])
plt.xlabel('')  # Removes x-axis label

# Add a legend
legend_handles = [plt.Line2D([0], [0], color='none', marker='s', markersize=10, 
                             markerfacecolor=color, markeredgecolor=color, linewidth=1.5, label=test) 
                  for test, color in zip(tests, edge_colors)]
plt.legend(handles=legend_handles, title="Tools", fontsize=10, title_fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust spines and save the plot
sns.despine(right=True, top=True, offset=10, trim=True)
#output_path = r"C:\Users\HP\Downloads\PBMC_INDEPENDENT_CCC.png"
#plt.savefig(output_path, dpi=1200, bbox_inches='tight')

# Display the plot
plt.tight_layout()
plt.show()
```
